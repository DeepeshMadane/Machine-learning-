```python
df_country_code = pd.read_excel('Country-Code.xlsx')
```

```python
df_country_code.head()
```

```python
final_df=pd.merge(df,df_country_code,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_count=final_df.Country.value_counts().values
```

```python
#top 3 countries using zomato
plt.pie(country_count[:3],labels=country_names[:3],autopct='%1.2f%%')
```
