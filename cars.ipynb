```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('cars.csv')
```

```python
dataset.head()
```

```python
dataset=dataset.drop(columns=['Car_ID'],axis=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
missing_value=dataset.isnull().sum()
```

```python
missing_value
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
categorical_columns = [column for column in dataset.columns if dataset[column].dtype=="O"]
```

```python
dataset[categorical_columns] = dataset[categorical_columns].apply(lambda x: le.fit_transform(x))
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
y = y.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler(copy=True, with_mean=True, with_std=True)
sc_y = StandardScaler(copy=True, with_mean=True, with_std=True)
X = sc_x.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, shuffle =True)
```

polynomial


```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=150,random_state=33)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
r2
```
